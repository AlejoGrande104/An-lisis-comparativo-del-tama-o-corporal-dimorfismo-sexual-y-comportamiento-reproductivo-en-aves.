Se busca evaluar las diferencias de tamaño corporal entre machos y hembras en aves y describir patrones generales de dimorfismo sexual a escala comparativa.



```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# Variables morfológicas por sexo
male_vars = ['M_mass', 'M_wing', 'M_tarsus', 'M_bill', 'M_tail']
female_vars = ['F_mass', 'F_wing', 'F_tarsus', 'F_bill', 'F_tail']

# Filtrar especies con datos completos para machos y hembras
df_sex = df[['Species_name'] + male_vars + female_vars].dropna()

# Construir variables de dimorfismo sexual (log-ratio)
ssd_data = pd.DataFrame()
ssd_data['Species_name'] = df_sex['Species_name']

for m, f in zip(male_vars, female_vars):
    ssd_data[m.replace('M_', 'SSD_')] = np.log(df_sex[m] / df_sex[f])

# Matriz de características (sin identificadores)
X = ssd_data.drop(columns='Species_name')
```

**ssd_data**: Para cada medida corporal, calculo cuánto más grande es el macho respecto a la hembra usando un cociente y lo paso a logaritmo. Eso me da una medida estándar de dimorfismo sexual: si da positivo, el macho es más grande; si da negativo, la hembra; y si da cerca de cero, son parecidos.

Con eso armo una tabla final que tiene solo esas diferencias entre sexos y saco el nombre de la especie, porque para el modelo lo que importa son los valores, no la etiqueta.

---








Se busca investigar si existe una relación entre el tamaño corporal de las aves y variables reproductivas como  son el tamaño del huevo y el tamaño de la puesta

```python

#Construir variables de tamaño corporal promedio

df['Mean_mass']   = (df['M_mass']   + df['F_mass'])   / 2
df['Mean_wing']   = (df['M_wing']   + df['F_wing'])   / 2
df['Mean_tarsus'] = (df['M_tarsus'] + df['F_tarsus']) / 2
df['Mean_bill']   = (df['M_bill']   + df['F_bill'])   / 2
df['Mean_tail']   = (df['M_tail']   + df['F_tail'])   / 2


#Seleccionar variables para el PCA de tamaño corporal


size_vars = [
    'Mean_mass',
    'Mean_wing',
    'Mean_tarsus',
   'Mean_bill',
    'Mean_tail'
]

df_size = df[['Species_name'] + size_vars].dropna()

X = df_size[size_vars]
```

Aca se necesita calcular el tamaño absoluto y no el relativo, tamaño relativo sirve mas para la forma, y lo que me interesa saber es la relacion huevo y morfologia

```python
# Normalización Min-Mam

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

df_size_normalized = pd.DataFrame(
    X_scaled,
    columns=size_vars,
    index=df_size['Species_name']
)


# PCA de tamaño corporal

pca = PCA(n_components=2)
X_pca = pca.fit_transform(df_size_normalized)

# Resultados del PCA
pca_df = pd.DataFrame({
    'standard_deviation': np.sqrt(pca.explained_variance_),
    'proportion_of_variance': pca.explained_variance_ratio_,
    'cumulative_proportion': np.cumsum(pca.explained_variance_ratio_)
})
pca_df.index = [f'PC{i}' for i in range(1, len(pca_df) + 1)]
pca_df.index.name = 'component'

print("Resumen del PCA de tamaño corporal:")
print(pca_df)
print("\n")
```

```python
# Loadings de las variables

loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i}' for i in range(1, pca.n_components_ + 1)],
    index=size_vars
)

print("Cargas de las variables en los componentes:")
print(loadings)
print("\n")



# Crear DataFrame final con PC1 y variables reproductivas
df_pca_scores = pd.DataFrame(
    X_pca,
    columns=['PC1_size', 'PC2_size'],
    index=df_size['Species_name']
)

df_repro = df[['Species_name', 'Egg_mass', 'Clutch_size']].dropna()

df_final = df_pca_scores.merge(
    df_repro,
    left_index=True,
    right_on='Species_name'
)



```

No alcanzan los colores para poder diferencias cada especie entonces se opto por agrupar las familias de aves en categorías taxonómicas amplias para reducir la dimensionalidad del conjunto de datos y facilitar comparaciones biológicamente interpretables

```python
import pandas as pd

# Función para agrupar
def get_grupo(fam):
    if fam in [1, 3, 4, 5]:                # Ratites y tinámidos (no voladores basales)
        return "Ratites / Tinamiformes"
    elif fam in [6, 7, 8, 9, 10]:          # Galliformes (faisanes, codornices, etc.)
        return "Galliformes"
    elif fam in [50, 51, 83, 84, 85]:      # Rapaces (búhos, águilas, halcones)
        return "Rapaces (Strigiformes + Accipitriformes + Falconiformes)"
    elif fam in [14, 12, 13, 93, 94, 95, 96, 97, 98, 99, 100]:  # Aves acuáticas / marinas
        return "Aves acuáticas / marinas"
    elif fam >= 101:                       # Paseriformes (la mayoría de las aves modernas)
        return "Paseriformes"
    else:
        return "Otros (no-paseriformes medios)"

# Aplica la agrupación
df_pca = pd.DataFrame({'PC1': x, 'PC2': y, 'Family': familia})
df_pca['Grupo'] = df_pca['Family'].apply(get_grupo)

# Gráfico
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Grupo',
                palette='Set1', alpha=0.8, s=60, edgecolor='gray')
plt.legend(title='Grupo evolutivo', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.xlabel('PC1 - Tamaño corporal general')
plt.ylabel('PC2 - Forma y dimorfismo sexual')
plt.title('PCA de medidas morfológicas en aves\nColoreado por grupo evolutivo (más claro y útil)')
plt.tight_layout()
plt.show()
```

**Para entender mejor el grafico...** A mayor PC1 implica un mayor tamaño de dicha especie. Un PC2 Alto significa Macho mucho más grande o con otras medidas muy distintas que la hembra, y uno cercano a  cero Macho y hembra casi iguales,y el negativo es revese hembra mas grande que el macho.

**Objetivo 4**: Clasificador de sistemas de emparejamiento. Monogano vs No monogano


---



```python
print(df['Mating_System'].value_counts())
```

Viva la monogamia!!! En el planeta cerca del 90 por ciento de las especies de aves son monogamas. Esta estrategia evolutiva es clave para asegurar la supervivencia de las crías, maximizando la eficiencia en la incubación y alimentación por parte de ambos padres

Pero para el clasificador la clase 2  sesga mucho el modelo asi que conviene eliminarla. Y tambien hacer una clasificacion binaria.

```python
# Filtrar solo filas con Mating_System válido (no NaN y no -999)
df_mating = df[df['Mating_System'].notna() & (df['Mating_System'] != -999)].copy()

# Eliminar clase 2
df_mating = df_mating[df_mating['Mating_System'] != 2]
```

```python
features = [
    'M_mass', 'F_mass',
    'M_wing', 'F_wing',
    'M_tarsus', 'F_tarsus',
    'M_bill', 'F_bill',
    'M_tail', 'F_tail'
]


# Estrictamente monogana solo es la primara clase, las 4 restantes son variaciones de la poliginia
def monogamy_binary(x):
    if x == 1:
        return 0  # Monógamo
    else:
        return 1  # No monógamo

df_mating['monogamy_class'] = df_mating['Mating_System'].apply(monogamy_binary)

X = df_mating[features]
y = df_mating['monogamy_class']

```

```python
# RANDOM FOREST CON OOB
rf_mono = RandomForestClassifier(
    n_estimators=2000,
    class_weight="balanced",
    oob_score=True,
    random_state=0,
    n_jobs=-1
)

rf_mono.fit(X, y)

print(f"OOB accuracy: {rf_mono.oob_score_:.3f}")

# PREDICCIONES OOB
y_oob_pred = np.argmax(rf_mono.oob_decision_function_, axis=1)
```

Mostrar resultados

```python

# Reporte de clasificacion

print("\nClasificación Monógamo vs No Monógamo (OOB)")
print(classification_report(
    y,
    y_oob_pred,
    target_names=['Monógamo', 'No monógamo']
))

# MATRIZ DE CONFUSIÓN

cm = confusion_matrix(y, y_oob_pred)

print("Matriz de confusión OOB:")
print(cm)

# HEATMAP

plt.figure(figsize=(5,4))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Monógamo', 'No monógamo'],
    yticklabels=['Monógamo', 'No monógamo']
)
plt.xlabel("Clase predicha")
plt.ylabel("Clase real")
plt.title("Matriz de Confusión OOB – Sistema de Apareamiento")
plt.tight_layout()
plt.show()
```

Las especies monógamas tienden a concentrarse en rangos de menor masa corporal, probablemente asociados a estrategias de cuidado biparental. Sin embargo tienen igual presiones de selección sexual extrema, solo que compiten por cortejo de ritual no por tamaño corporal. Machos más coloridos-vistosos y complejas exhibiciones de cortejo

**CONCLUSION**

A pesar del fuerte desbalance entre clases, el modelo muestra un buen desempeño general y es capaz de identificar correctamente una proporción considerable de especies monógamas, sin colapsar hacia la clase mayoritaria. El uso de validación OOB sugiere que los resultados son estables y no producto de sobreajuste.

**Objetivo 5**: Clasificador deacuerdo a la complejidad del ritual


---



El objetivo es predecir la complejidad del display (simple / intermedio / complejo) a partir del tamaño corporal y el dimorfismo sexual de la especie.

```python
features = [
    'M_mass', 'F_mass',
    'M_wing', 'F_wing',
    'M_tarsus', 'F_tarsus',
    'M_bill', 'F_bill',
    'M_tail', 'F_tail'
]
# Crear df_disp: solo filas con Display válido y todas las features completas
df_disp = df[
    df['Display'].notna() & (df['Display'].isin([1, 2, 3, 4, 5]))  # solo 1 a 5
].copy()
```

```python


# Display
df_disp = df_disp[df_disp['Display'].isin([1, 2, 3, 4, 5])]

# Se separa en 3 clases principales para no complejizar el modelo
def display_3clases(x):
    if x <= 2:
        return 0  # simple
    elif x == 3:
        return 1  # intermedio
    else:
        return 2  # complejo

df_disp['disp_3class'] = df_disp['Display'].apply(display_3clases)

X = df_disp[features]
y = df_disp['disp_3class']
```

 Si dejo las  5 clases separadas, el modelo se vuelve inestable: no aprende bien las clases minoritarias y el rendimiento cae mucho
 Al Agrupar en 3 clases (≤2 = simple, 3 = intermedio, >3 = complejo) reduce ruido y aumenta el número de ejemplos por clase, haciendo el problema mucho más tratable. Ya se habia probado con las 5 claes y asi se mejoro los resultados

```python
# RANDOM FOREST CON OOB
rf_disp = RandomForestClassifier(
    n_estimators=2000,
    class_weight="balanced",
    oob_score=True,
    random_state=0,
    n_jobs=-1
)

rf_disp.fit(X, y)
```

```python
print(f"OOB accuracy: {rf_disp.oob_score_:.3f}")
```

resultados

balancear los datos

```python
# PREDICCIONES OOB
y_oob_pred = np.argmax(rf_disp.oob_decision_function_, axis=1)

# CLASSIFICATION REPORT
print("\nDisplay simple / intermedio / complejo (OOB)")
print(classification_report(y, y_oob_pred))

# MATRIZ DE CONFUSIÓN
cm = confusion_matrix(y, y_oob_pred)
print("Matriz de confusión OOB:")
print(cm)

# HEATMAP DE MATRIZ
plt.figure(figsize=(5,4))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Simple', 'Intermedio', 'Complejo'],
    yticklabels=['Simple', 'Intermedio', 'Complejo']
)
plt.xlabel("Clase predicha")
plt.ylabel("Clase real")
plt.title("Matriz de Confusión OOB")
plt.tight_layout()
plt.show()
```

**CONCLUSION:**

El modelo tiene una visión muy “aplanada” del problema: distingue bien lo simple, pero le cuesta mucho escalar en complejidad. Intermedio y complejo quedan absorbidos por la clase simple, lo que sugiere que el principal desafío no es mejorar la detección de lo simple, sino darle al modelo mejores señales para entender y separar los niveles de complejidad más altos.

# Conclusiones finales

Este trabajo muestra que el dimorfismo sexual en aves no es algo universal, sino que depende mucho del tipo de ave y de su historia evolutiva. No todas las especies siguen el mismo patrón: en algunos grupos los machos son más grandes o más llamativos, en otros pasa lo contrario, y eso tiene sentido cuando mirás para qué sirve cada rasgo.

Las diferencias entre machos y hembras están influenciadas por dos fuerzas distintas:

**La selección sexual,** que exagera rasgos visibles (como cola, alas u ornamentos) cuando los machos compiten o atraen pareja.

**La selección natural**, que puede favorecer a las hembras en rasgos más funcionales, por ejemplo cuando necesitan ser más eficientes para reproducirse o sobrevivir.

El tamaño corporal, en cambio, parece responder más a cuestiones ecológicas que sexuales. Los modelos de clasificación refuerzan estas ideas, ya que logran distinguir en parte estos patrones, aunque su desempeño es peor en los grupos menos representados, lo que muestra que el desbalance de datos es una limitación importante.

En conjunto, los resultados aportan evidencia cuantitativa de que el dimorfismo sexual en aves surge de la combinación entre ecología, selección sexual y herencia evolutiva, y muestran que para estudiar estos procesos es clave usar enfoques multivariados y bases de datos bien balanceadas.
